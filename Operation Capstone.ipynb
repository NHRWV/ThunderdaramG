```python

from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords
import nltk
from nltk import word_tokenize
from nltk.stem.porter import PorterStemmer
import re
from nltk.stem import WordNetLemmatizer




MSG = ["script/Gladiator.txt",
       "script/Casablanca.txt", 
       "script/Notting Hill.txt",
       "script/Pride and Prejudice.txt"
      ]
MSB = ["script/Jack and Jill.txt", 
       "script/John Carter.txt", 
       "script/Valerian and the City of a Thousand Planets.txt", 
       "script/Aloha.txt"]
    
def movie_script(script): # 대상이 되는 대본 읽고 토크나이즈
    with open(script, "r", encoding = "utf-8") as file:
        data = file.read()
        
    t = RegexpTokenizer("\w+")
    tokens_en = t.tokenize(data)
        
    return tokens_en
```

```python
len(movie_script(MSG[1]))
```

```python
def merge_script_1(func): # 해당하는 영화 리스트 안에 있는 모든 대본 합치기 ver.1
    merged = []
    for i in range(len(MSG)):
        merged.append(movie_script(func[i]))
    return merged

def merge_script_2(func): # 해당하는 영화 리스트 안에 있는 모든 대본 합치기 ver.2
    merged = []
    for i in range(len(MSG)):
        merged += movie_script(func[i])
    return merged
```

```python
len(merge_script(MSG))
```

```python
stopwords_list = stopwords.words("english")
wl = WordNetLemmatizer()
stopwords_list.extend(["bennet", "jane", "collins", "catherine", "darcy", "charlotte", "maximus", "lizzie", "laszlo", "elizabeth", "wickham", "bingley", "lydia"])

def tokenize(text):
    min_length = 3 
    # 소문자화
    words = [a.lower() for a in text]
    words = [word for word in words if word not in stopwords_list] # 불용어 제거.
    ##tokens = [PorterStemmer().stem(token) for token in words] 이건 어간추출. 
    tokens = [wl.lemmatize(w) for w in words] # 요거는 변형된 형태의 단어 기본 형태로 복원.
    
    p = re.compile('[a-zA-Z]+'); # 원하는 정규표현식 특징을 변수에 할당. 알파벳만 가져온다. 
    # 그런데 이미 모든 문자를 소문자화 시킨 상태에서 A-Z는 필요없지 않나??? 
    
    filtered_tokens = list(filter (lambda token: p.match(token) and len(token) >= min_length, tokens))
    
    return filtered_tokens 
```

```python
msg = merge_script_1(MSG)
msb = merge_script_1(MSB)

msg_tokenized_gen = [tokenize(lines) for lines in msg]
msb_tokenized_gen = [tokenize(lines) for lines in msb]

msg_tokenized = tokenize(merge_script_2(MSG))
msb_tokenized = tokenize(merge_script_2(MSB))

```

```python
from gensim.corpora import Dictionary
# 딕셔너리 만들기...bag of words
msg_dictionary = Dictionary(msg_tokenized_gen)
msb_dictionary = Dictionary(msb_tokenized_gen)
#문서안의 각 단어에 대해서 유니크한 ID를 만든다. 


# Filter out words that occur less than 10 documents, or more than 20% of the documents.
msg_dictionary.filter_extremes(no_below=2)
msb_dictionary.filter_extremes(no_below=2)
print('Number of unique words after removing rare and common words:', len(msg_dictionary))
```

```python
print(msg_tokenized_gen)
```

```python
msg_corpus = [msg_dictionary.doc2bow(doc) for doc in msg_tokenized_gen] # 코퍼스
msb_corpus = [msb_dictionary.doc2bow(doc) for doc in msb_tokenized_gen]
```

```python
from gensim.models import LdaModel

# Make a index to word dictionary.
msg_temp = msg_dictionary[0]# 사전 만든 것 불러오려면 필요.. 없으면 에러발생.
msb_temp = msb_dictionary[0]
msg_id2word = msg_dictionary.id2token
msb_id2word = msb_dictionary.id2token



# pass & iteration 은 최대한 많은게 좋다!
def gensimLDA(corpus_, id2word_, topic_num, chunksize_, passes_, iterations_, eval_every_):
    model = LdaModel(corpus = corpus_,
                     id2word = id2word_,
                     alpha = 'auto', eta = 'auto',
                     iterations = iterations_,
                     num_topics = topic_num, 
                     passes = passes_, 
                     eval_every = eval_every_, 
                     chunksize = chunksize_)
    return model
    
msg_model = gensimLDA(msg_corpus, msg_id2word, 10, 100, 20, 100, 1)
msb_model = gensimLDA(msb_corpus, msb_id2word, 10, 100, 20, 100, 1)
```
